```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, ServiceContext, StorageContext, load_index_from_storage
from llama_index.llms.palm import PaLM

from langchain.embeddings.huggingface import HuggingFaceBgeEmbeddings
from llama_index.core import Settings

from IPython.display import display, Markdown
```

```python
data =  "/content/gdrive/MyDrive/Colab Notebooks/llama_index_pdf_reader"
```

```python
documents = SimpleDirectoryReader(data).load_data()
```

## Split the text into chunks


```python
import google.generativeai as genai

# Used to securely store your API key
from google.colab import userdata

GOOGLE_API_KEY=userdata.get('Bard-API')

genai.configure(api_key=GOOGLE_API_KEY)
```

```python
llm = PaLM(api_key=GOOGLE_API_KEY)
embed_model = HuggingFaceBgeEmbeddings(model_name="BAAI/bge-base-en")
```

```python
Settings.llm = llm
Settings.embed_model = embed_model
#Settings.node_parser = SentenceSplitter(chunk_size=512, chunk_overlap=64)
#Settings.num_output = 512
#Settings.context_window = 3900
Settings.chunk_size=512
Settings.chunk_overlap=64

index = VectorStoreIndex.from_documents(documents)
```

## **Store  and loading the index**


```python
index.storage_context.persist()
```

```python
query_engine = index.as_query_engine()
```

```python
query = "rivers in india in 500 words"
response = query_engine.query(query)
print(response)
```

```python
display(Markdown(f"{response}"))
```

## Vector DB - FAISS


```python
from langchain_community.document_loaders import PyPDFLoader

docs = PyPDFLoader('/content/gdrive/MyDrive/Colab Notebooks/llama_index_pdf_reader/ind_geo.pdf').load()
print(docs)
```

```python
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=512, chunk_overlap=64)
texts = text_splitter.split_documents(docs)
texts[:5]
```

```python
db = FAISS.from_documents(texts, embed_model)
```

```python
query = "rivers in india in 500 words"
result =db.similarity_search(query)
result
```
